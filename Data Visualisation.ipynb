```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,11)
y=x*2
```

```python
#plot function : line plot

plt.plot(x,y)
plt.show()
```

```python
# Adding Title and labels to your graph
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x-Label")
plt.ylabel("y-Label")
plt.show()
```

```python
# Adding colors and line style to your graph

plt.plot(x,y,color='red',linestyle=":")
plt.title("Line Plot")
plt.xlabel("x-Label")
plt.ylabel("y-Label")
plt.show()
```

```python
# Formating line width: specify linewitdh attribute

plt.plot(x,y,color='red',linestyle=":",linewidth=4)
plt.title("Line Plot")
plt.xlabel("x-Label")
plt.ylabel("y-Label")
plt.show()
```

```python
# Adding grid to your graph

plt.plot(x,y,color='red',linestyle=":",linewidth=4)
plt.title("Line Plot")
plt.xlabel("x-Label")
plt.ylabel("y-Label")
plt.grid(True)
plt.show()
```

```python
# Ploting two or more lines in single graph

z=x*3
plt.plot(x,y,color='blue',linestyle='-.')
plt.plot(x,z,color='green',linestyle=':')
plt.title("Two line in single graph")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid(True)
plt.show()
#similarly you can add n no. of lines in graph. We should try to make functions if no. of variables increases to more than 4
```

```python
# Subploting : creating two plots at same time
# we require more than 1 variables to plot 2 graphs at same time

x
y
z
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,color='r')

plt.subplot(1,2,2)
plt.plot(x,z,color='b')

plt.show()
```

```python
#we can also specify the size of the graph 
#figure function

plt.figure(figsize=(10,7))

plt.subplot(1,2,1)
plt.plot(x,y,color='r')

plt.subplot(1,2,2)
plt.plot(x,z,color='b')

plt.show()
```

```python
# Bar Plot
# used for categorical data

name=["Bob","Martin","Jhon"]

age=[34,56,86]

plt.bar(name,age,color='Red')
plt.title("Distribution of age")
plt.xlabel('Names')
plt.ylabel('Age')
plt.show()
```

```python
# Horizontal bar Graph

name=["Bob","Martin","Jhon"]

age=[34,56,86]

plt.barh(name,age,color='Blue')
plt.title("Distribution of age")
plt.xlabel('Names')
plt.ylabel('Age')
plt.show()
```

```python
# To add data labels we can make a function

def datalabels(a,b):
    for i in range(len(a)):
        plt.text(i,b[i],b[i])
```

```python
plt.bar(name,age,color='Blue')
datalabels(name,age)
plt.title("Distribution of age")
plt.xlabel('Names')
plt.ylabel('Age')
plt.show()
```

```python
# Scatter Plots: to check relationship between two or more variables

x=np.random.randint(10,100,10)
y=np.random.randint(20,50,10)

plt.scatter(x,y)
plt.show()
```

```python
# Specifying marker type and color and size of the marker

plt.scatter(x,y,marker="*",color="Green",s=100)
plt.title("Scatter Plot")
plt.xlabel("x-Label")
plt.ylabel("y-Label")
plt.show()
```

```python
# Two variable scatter plot and adding legends

x=np.random.randint(10,100,10)
y=np.random.randint(20,50,10)
z=np.random.randint(20,100,10)

plt.scatter(x,y,marker="*",color="Green",s=100)
plt.scatter(x,z,marker="^",color="Red",s=150)
plt.title("Scatter Plot")
plt.xlabel("x-Label")
plt.ylabel("y-Label")
plt.legend(["Green","Red"])
plt.show()
```

```python
# Histogram

# if you want to check frequency of data

data=np.random.randint(1,10,20)


plt.hist(data)

plt.show()

```

```python
# Box plot : used to check outlier and Inetr Quartile Range

x=np.random.randint(1,20,10)
y=np.random.randint(1,5,10)
z=np.random.randint(4,9,10)

plt.figure(figsize=(10,7))
data=list([x,y,z])

plt.boxplot(data)
plt.show()
```

```python
# Pie Chart: Used to view the frequency percentage of different categorical value

fruits=['Apple','Orange','Grapes']
quantity=[100,29,76]

plt.pie(quantity,labels=fruits,autopct='%0.1f%%')
plt.show()
```

```python
# Error Bar: Difference between actual data point and predicted data points

x=[1,2,3,4,5]
y=[4,2,1,4,3]

plt.errorbar(x,y,yerr=0.5)
plt.show()
```

```python
# Area Plot : 

week=[1,2,3,4,5]

sf1=[45,76,87,12,85]
sf2=[76,56,92,18,65]
sf3=[38,71,40,47,30]
```

```python
plt.stackplot(week,sf1,sf2,sf3)
plt.legend(['sf1','sf2','sf3'])
plt.show()
```
